### 🔗 Open Dataset Link & Advanced Analytics
Run the cell below to load the simulated Tamil Nadu public health dataset and generate advanced analytics on resource allocation and demographics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set seaborn style for better aesthetics
sns.set_theme(style="whitegrid")

print("🔗 NALAM AI - OPEN DATASET LINK")
print("======================================")
print("📥 Tamil Nadu Public Health Outbreak Dataset (Simulated):")
print("🔗 Link: https://raw.githubusercontent.com/gan1014/Nalam_AI_2.0/main/data/tn_disease_outbreak_2024.csv")
print("(Note: In this demo, we are using a synthetically generated dataset to ensure patient privacy compliance.)\n")

# --- DATA GENERATION ---
districts = ["Chennai", "Coimbatore", "Madurai", "Salem", "Tirunelveli"]
diseases = ["Dengue", "Scrub Typhus", "Gastroenteritis", "Heat Stroke"]

# Create a heatmap dataset
heatmap_data = pd.DataFrame({
    "Dengue": [245, 89, 120, 65, 40],
    "Scrub Typhus": [12, 45, 23, 80, 15],
    "Gastroenteritis": [310, 150, 180, 90, 110],
    "Heat Stroke": [45, 20, 60, 110, 85]
}, index=districts)

# Create resource allocation dataset
resource_data = pd.DataFrame({
    "Resource": ["Oxygen Cylinders", "IV Fluids (Liters)", "Dengue NS1 Kits", "Hospital Beds"],
    "Currently Available": [450, 1200, 300, 850],
    "Required (Next 7 Days)": [800, 2500, 1000, 1200]
})

# --- PLOT 1: Disease Distribution Heatmap ---
fig = plt.figure(figsize=(16, 12))

# Subplot 1: Heatmap
ax1 = plt.subplot(2, 2, 1)
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlOrRd", linewidths=.5, ax=ax1)
ax1.set_title("🗺️ Regional Disease Outbreak Density", fontsize=14, pad=15, fontweight="bold")
ax1.set_ylabel("Districts", fontweight="bold")

# --- PLOT 2: Resource Allocation Gap (Bar Chart) ---
ax2 = plt.subplot(2, 2, 2)
x = np.arange(len(resource_data["Resource"]))
width = 0.35

ax2.bar(x - width/2, resource_data["Currently Available"], width, label="Available Stock", color="#10b981")
ax2.bar(x + width/2, resource_data["Required (Next 7 Days)"], width, label="Predicted Requirement", color="#ef4444")

ax2.set_ylabel("Unit Count", fontweight="bold")
ax2.set_title("🏥 Medical Resource Prediction vs Availability", fontsize=14, pad=15, fontweight="bold")
ax2.set_xticks(x)
ax2.set_xticklabels(resource_data["Resource"], rotation=15)
ax2.legend()

# --- PLOT 3: Affected Patient Demographics (Pie Chart) ---
ax3 = plt.subplot(2, 2, 3)
age_groups = ["Children (0-12)", "Teens (13-19)", "Adults (20-55)", "Seniors (55+)"]
cases_by_age = [25, 15, 45, 15]
colors = ["#f472b6", "#38bdf8", "#fbbf24", "#c084fc"]
explode = (0.1, 0, 0, 0)  # highlight children

ax3.pie(cases_by_age, explode=explode, labels=age_groups, colors=colors, autopct="%1.1f%%",
        shadow=True, startangle=140, textprops={"fontweight": "bold"})
ax3.set_title("👥 Patient Demographics (Dengue)", fontsize=14, pad=15, fontweight="bold")

# --- PLOT 4: 30-Day Recovery vs Admission Trend ---
ax4 = plt.subplot(2, 2, 4)
days = np.arange(1, 31)
admissions = 50 + np.sin(days/3) * 20 + days * 2
recoveries = 30 + np.sin((days-3)/3) * 15 + days * 1.8

ax4.plot(days, admissions, label="New Admissions", color="#ef4444", linewidth=2.5)
ax4.plot(days, recoveries, label="Patient Recoveries", color="#10b981", linewidth=2.5)
ax4.fill_between(days, admissions, recoveries, where=(admissions > recoveries), color="#ef4444", alpha=0.1, label="Hospital Capacity Strain")

ax4.set_title("⚕️ Hospital Admissions vs Recoveries", fontsize=14, pad=15, fontweight="bold")
ax4.set_xlabel("Days", fontweight="bold")
ax4.set_ylabel("Number of Patients", fontweight="bold")
ax4.legend()

plt.tight_layout(pad=3.0)
plt.show()
```
