```python
# 5. Configure Environment Variables
import os

# !!! IMPORTANT: SET YOUR KEYS BELOW !!!
os.environ['HUGGINGFACE_API_KEY'] = 'YOUR_HUGGINGFACE_KEY_HERE'
os.environ['SUPABASE_URL'] = 'YOUR_SUPABASE_URL_HERE'
os.environ['SUPABASE_KEY'] = 'YOUR_SUPABASE_KEY_HERE'

with open('.env.local', 'w') as f:
    f.write(f"HUGGINGFACE_API_KEY={os.environ['HUGGINGFACE_API_KEY']}\n")
    f.write(f"NEXT_PUBLIC_SUPABASE_URL={os.environ['SUPABASE_URL']}\n")
    f.write(f"NEXT_PUBLIC_SUPABASE_ANON_KEY={os.environ['SUPABASE_KEY']}\n")
    f.write("ENVIRONMENT=production\n")
```

```python
# 6. Run the Server & Expose it
import subprocess
import threading

def run_server():
    subprocess.run(["npm", "run", "dev"])

threading.Thread(target=run_server).start()

print("Waiting for server to initialize...")
!sleep 10

print("\n--- YOUR PUBLIC URL IS BELOW ---")
!npx localtunnel --port 3000
```
